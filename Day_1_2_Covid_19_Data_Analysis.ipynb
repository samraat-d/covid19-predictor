```
path = "https://raw.githubusercontent.com/umangkejriwal1122/Machine-Learning/master/Data%20Sets/covid_19_clean_complete.csv"
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```
data = pd.read_csv(path)
data
```

```
data.info()
```

```
data = pd.read_csv(path,parse_dates=["Date"])
data
#### to convert date to numeric value
```

```
data.info()
```

```
data.drop(['Province/State'],axis =1,inplace = True)
data.head()
```

```
data.rename(columns={"Country/Region":"Country"},inplace=True)
data.head()
```

```
#### Print the latest data
top = data[data['Date']==data['Date'].max()]
top
```

```
#### Grouping data by country as each country has many province
#### All data will be merged while grouping
world = top.groupby('Country')["Confirmed","Deaths","Active","Recovered"].sum()
world.head()
```

```
world.reset_index(inplace = True)
world.head()
#### This is to make index coloumn as country was before the unique index
```

```
world[world['Country']=='China']
#### checks data only for China
```

```
#### Plot the data on World Map (Death Cases)
world_map = px.choropleth(world,locations="Country",locationmode="country names"
                          ,color_continuous_scale = "Peach",color = "Deaths",
                          range_color = [1,20000],
                          title="Death Cases in Different Countries")
world_map.show()
#### plotting for country names by comparing to actual country name, if data
#### is correct then it will plot data
#### for colour = deaths, this can be changed to active
#### range is division of colour by 20,000 samples
```

```
#### Plot confirmed cases vs date graph
total_confirmed_cases = data.groupby("Date")['Confirmed'].sum().reset_index()
total_confirmed_cases
```

```
plt.figure(figsize=(20,8)) # figure size
plt.xticks(rotation=90,fontsize=8) # x labels rotation
plt.title("Confirmed Cases Vs Date Graph")
sns.pointplot(x=total_confirmed_cases['Date'].dt.date,                                             # removes time
              y=total_confirmed_cases['Confirmed'],color='red')
plt.xlabel("Dates --->>",fontsize = 12)
plt.ylabel("Total Confirmed Cases --->>",fontsize = 12)
plt.show()
```

```
#### Plot recovered/confirmed cases vs date graph
total_recovered_cases = data.groupby("Date")['Recovered'].sum().reset_index()
total_recovered_cases
```

```
plt.figure(figsize=(22,8))
plt.xticks(rotation=90,fontsize=8)
plt.title("Confirmed/Recovered Cases Vs Date Graph")
sns.pointplot(x=total_confirmed_cases['Date'].dt.date,
              y=total_confirmed_cases['Confirmed'],color='red')
sns.pointplot(x=total_recovered_cases['Date'].dt.date,
              y=total_recovered_cases['Recovered'],color='green')
plt.xlabel("Dates --->>",fontsize = 12)
plt.ylabel("Total Confirmed/Recovered Cases --->>",fontsize = 12)
plt.show()
```

```
#### Top 20 Countries Having Maximum Confirmed Cases
top_cases = top.groupby("Country")["Confirmed","Recovered"].sum()
top_cases.head()
```

```
top_cases = top_cases.sort_values("Confirmed",ascending=False).reset_index()
top_cases
```

```
top_20 = top_cases.head(20)
top_20
```

```
plt.figure(figsize=(12,6))
sns.barplot(x=top_20['Confirmed'],y=top_20['Country'],
            color='red',label="Confirmed")
sns.barplot(x=top_20['Recovered'],y=top_20['Country'],
            color='green',label="Recovered")
plt.legend(loc=4)
# loc is the quadrant name or location of labels
plt.show()
```

```
india = data[data['Country']=='India']
india
```

```
india_confirmed = india.groupby('Date')['Confirmed'].sum().reset_index()
india_confirmed
```

```
china = data[data['Country']=='China']
china_confirmed = china.groupby('Date')['Confirmed'].sum().reset_index()
china_confirmed
```

```
# point graph of two countries
plt.figure(figsize=(20,8)) # figure size
plt.xticks(rotation=90,fontsize=8) # x labels rotation
plt.title("Confirmed Cases Vs Date Graph")
sns.pointplot(x=india_confirmed['Date'].dt.date,
              y=india_confirmed['Confirmed'],color='red')
sns.pointplot(x=china_confirmed['Date'].dt.date,
              y=china_confirmed['Confirmed'],color='blue')
plt.xlabel("Dates --->>",fontsize = 12)
plt.ylabel("Total Confirmed Cases --->>",fontsize = 12)
plt.show()
```

```
# line graph with legend of two countries
plt.figure(figsize=(10,8)) # figure size
plt.xticks(rotation=90,fontsize=8) # x labels rotation
plt.title("Confirmed Cases Vs Date Graph")
plt.plot(india_confirmed['Date'].dt.date,
              india_confirmed['Confirmed'],color='red',label="India")
plt.plot(china_confirmed['Date'].dt.date,
              china_confirmed['Confirmed'],color='blue',label="India")
plt.xlabel("Dates --->>",fontsize = 12)
plt.ylabel("Total Confirmed Cases --->>",fontsize = 12)
plt.legend(loc=2)
plt.show()
```
